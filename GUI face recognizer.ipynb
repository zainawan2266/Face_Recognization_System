```python
import tkinter as tk
from tkinter import messagebox
import cv2
import os
import re
from PIL import Image
import numpy as np

# Hover effects
def on_enter(e):
    e.widget.configure(bg='#1abc9c')

def on_leave(e):
    e.widget.configure(bg='#16a085')

def on_enter2(e):
    e.widget.configure(bg='#3498db')

def on_leave2(e):
    e.widget.configure(bg='#2980b9')

def on_enter3(e):
    e.widget.configure(bg='#e67e22')

def on_leave3(e):
    e.widget.configure(bg='#d35400')

# Main window
window = tk.Tk()
window.title("Face Recognition System")
window.configure(bg="#2c3e50")
window.geometry("900x600")
window.resizable(True, True)

# Center frame
main_frame = tk.Frame(window, bg="#2c3e50")
main_frame.place(relx=0.5, rely=0.5, anchor="center")

# Fonts
heading_font = ("Segoe UI", 24, "bold")
label_font = ("Segoe UI", 14)
entry_font = ("Segoe UI", 13)
button_font = ("Montserrat", 12, "bold")

# Heading
tk.Label(main_frame, text="Face Recognition System", font=heading_font, fg="white", bg="#2c3e50").grid(column=0, row=0, columnspan=3, pady=(0, 29))

# Entry styles
style = {
    "bd": 2,
    "relief": "solid",
    "highlightthickness": 1,
    "highlightbackground": "#7f8c8d",
    "highlightcolor": "#95a5a6",
    "font": entry_font,
    "width": 40
}

# Labels and Entries
tk.Label(main_frame, text="Name", font=label_font, fg="white", bg="#2c3e50").grid(column=0, row=1, sticky="e", padx=10, pady=5)
t1 = tk.Entry(main_frame, **style)
t1.grid(column=1, row=1, padx=10, pady=5)

tk.Label(main_frame, text="Age", font=label_font, fg="white", bg="#2c3e50").grid(column=0, row=2, sticky="e", padx=10, pady=5)
t2 = tk.Entry(main_frame, **style)
t2.grid(column=1, row=2, padx=10, pady=5)

tk.Label(main_frame, text="Address", font=label_font, fg="white", bg="#2c3e50").grid(column=0, row=3, sticky="e", padx=10, pady=5)
t3 = tk.Entry(main_frame, **style)
t3.grid(column=1, row=3, padx=10, pady=5)

# Button style
button_style = {
    "font": button_font,
    "padx": 25,
    "pady": 10,
    "bd": 0,
    "relief": "flat",
    "fg": "white",
    "width": 15,
    "cursor": "hand2"
}

# Get next user ID
def get_next_user_id(data_dir="data"):
    if not os.path.exists(data_dir):
        return 1
    existing_files = os.listdir(data_dir)
    user_ids = [int(re.match(r"user\.(\d+)\.", file).group(1)) for file in existing_files if re.match(r"user\.(\d+)\.", file)]
    return max(user_ids) + 1 if user_ids else 1

# Generate dataset
def generate_dataset():
    if t1.get() == "" or t2.get() == "" or t3.get() == "":
        messagebox.showinfo('Result', 'Please! Provide complete details of the user')
        return

    face_classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

    def face_cropped(img):
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_classifier.detectMultiScale(gray, 1.3, 5)
        for (x, y, w, h) in faces:
            return img[y:y+h, x:x+w]
        return None

    if not os.path.exists("data"):
        os.makedirs("data")

    user_id = get_next_user_id()
    cap = cv2.VideoCapture(1)
    if not cap.isOpened():
        cap = cv2.VideoCapture(0)
        if not cap.isOpened():
            messagebox.showerror("Error", "Webcam not detected")
            return

    img_id = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            continue
        cropped = face_cropped(frame)
        if cropped is not None:
            img_id += 1
            face = cv2.resize(cropped, (200, 200))
            face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
            cv2.imwrite(f"data/user.{user_id}.{img_id}.jpg", face)
            cv2.imshow("Cropped face", face)
        if cv2.waitKey(1) == 13 or img_id == 200:
            break

    cap.release()
    cv2.destroyAllWindows()
    messagebox.showinfo('Result', 'Generate Dataset Completed!')

# Train classifier
def train_classifier():
    data_dir = "data"
    if not os.path.exists(data_dir):
        messagebox.showerror("Error", "No dataset found! Generate it first.")
        return

    path = [os.path.join(data_dir, f) for f in os.listdir(data_dir) if f.endswith(".jpg")]
    faces, ids = [], []

    for image in path:
        try:
            img = Image.open(image).convert('L')
            imageNp = np.array(img, 'uint8')
            id = int(os.path.split(image)[1].split(".")[1])
            faces.append(imageNp)
            ids.append(id)
        except Exception as e:
            print(f"Skipping file {image}, error: {e}")

    clf = cv2.face.LBPHFaceRecognizer_create()
    clf.train(faces, np.array(ids))
    clf.write("classifier.xml")
    messagebox.showinfo('Result!', 'Training data set completed')

# Detect face
def detect_face():
    def draw_boundary(img, classifier, scaleFactor, minNeighbors, color, text, clf):
        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        features = classifier.detectMultiScale(gray_img, scaleFactor, minNeighbors)
        for (x, y, w, h) in features:
            cv2.rectangle(img, (x, y), (x + w, y + h), color, 2)
            id, pred = clf.predict(gray_img[y:y + h, x:x + w])
            confidence = int(100 * (1 - pred / 300))
            if confidence > 70:
                name = "unknown" if id == 1 else "Ali" if id == 2 else str(id)
                cv2.putText(img, name, (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 1)
            else:
                cv2.putText(img, "Zain", (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 1)
        return img

    faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
    clf = cv2.face.LBPHFaceRecognizer_create()
    clf.read("classifier.xml")
    video_capture = cv2.VideoCapture(0)

    while True:
        ret, img = video_capture.read()
        if not ret or img is None:
            continue
        img = draw_boundary(img, faceCascade, 1.3, 6, (255, 255, 255), "Face", clf)
        if img is not None:
            cv2.imshow("face Detection", img)
        if cv2.waitKey(1) == 13:
            break

    video_capture.release()
    cv2.destroyAllWindows()

# Buttons
b1 = tk.Button(main_frame, text="Training", bg="#16a085", command=train_classifier, **button_style)
b1.grid(column=0, row=5, padx=5, pady=10)
b1.bind("<Enter>", on_enter)
b1.bind("<Leave>", on_leave)

b2 = tk.Button(main_frame, text="Detect Faces", bg="#2980b9", command=detect_face, **button_style)
b2.grid(column=1, row=5, padx=5, pady=20)
b2.bind("<Enter>", on_enter2)
b2.bind("<Leave>", on_leave2)

b3 = tk.Button(main_frame, text="Generate Dataset", bg="#d35400", command=generate_dataset, **button_style)
b3.grid(column=2, row=5, padx=5, pady=30)
b3.bind("<Enter>", on_enter3)
b3.bind("<Leave>", on_leave3)

# Responsive layout
main_frame.grid_columnconfigure(0, weight=1)
main_frame.grid_columnconfigure(1, weight=1)
main_frame.grid_columnconfigure(2, weight=1)

# Start GUI
window.mainloop()
```
